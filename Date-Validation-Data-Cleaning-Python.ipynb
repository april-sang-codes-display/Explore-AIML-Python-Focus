```python
#A Python program that reads and validates dates

from datetime import date
import calendar

strdate = input("Enter a date in format yy.mm.dd: ")

while strdate != "STOP":
    # Extract day, month, year values
    vals = strdate.split(".")
    day = int(vals[2])
    month = int(vals[1])
    year = int(vals[0])

    # Handle Y2K problem by assuming if year > 60 in 1900s
    if year > 60:
        year = year + 1900
    else:
        year = year + 2000

    # Convert month number to month name
    if month == 1:
        monthname = "January"
    elif month == 2:
        monthname = "February"
    elif month == 3:
        monthname = "March"
    elif month == 4:
        monthname = "April"
    elif month == 5:
        monthname = "May"
    elif month == 6:
        monthname = "June"
    elif month == 7:
        monthname = "July"
    elif month == 8:
        monthname = "August"
    elif month == 9:
        monthname = "September"
    elif month == 10:
        monthname = "October"
    elif month == 11:
        monthname = "November"
    elif month == 12:
        monthname = "December"

    # Print date in new format
    print("Date:", str(year)+",", monthname, day)

    # Check if valid number of days. Note: Not handling leap years.
    if month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12:
        if day > 31:
            print("Error: Invalid date")      
    elif month == 2:
        if day > 28:
            print("Error: Invalid date")
    else:
        if day > 30:
            print("Error: Invalid date")

    #Convert date into a Python date
    pydate = date(year, month, day)
    print(pydate)

    #Convert month number to name using Python module
    print(calendar.month_name[month])
    strdate = input("Enter a date in format yy.mm.dd: ")
```
