## MultiIndexing (hierarchical indexing) performed via Pivot tables. 

- Make a dataframe with hierarchical indexing in the columns. Hint: Use .pivot_table(). 

- Write a program that returns all values in the MultiIndex (name of inner levels). For example, for Age, it should return 'From 15 to 24 years', etc. 

```python
import numpy as np
import pandas as pd

data = pd.read_csv("employee.csv")

table = pd.pivot_table(data,index=['DATE'],columns=['UNIT','AGE','SEX','INDIC_EM','GEO'],values=['Value'])
table
```

```python
data = pd.read_csv("employee.csv")
data

age = pd.pivot_table(data,index=['AGE'])
print(age.index.values)

sex = pd.pivot_table(data,index=['SEX'])
print(sex.index.values)

geo = pd.pivot_table(data,index=['GEO'])
print(geo.index.values)

indic = pd.pivot_table(data,index=['INDIC_EM'])
print(indic.index.values)

unit = pd.pivot_table(data,index=['UNIT'])
print(unit.index.values)
```
